```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import random
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import scipy.io.wavfile as sci_wav  # Open wav files
import os  # Manipulate files
import IPython
```

```python
def read_wav_files(wav_files):
    '''Returns a list of audio waves
    Params:
        wav_files: List of .wav paths

    Returns:
        List of audio signals
    '''
    if not isinstance(wav_files, list):
        wav_files = [wav_files]
    return [sci_wav.read(ROOT_DIR + f)[1] for f in wav_files]
```

```python
import os
# List the wav files
ROOT_DIR = './cats_dogs/'
data_path = os.listdir(ROOT_DIR)
y = [0 if 'cat' in f else 1 for f in data_path]  # Labels, 0-> Cats 1-> Dogs
```

```python
a = read_wav_files(data_path)
```

```python
fig, axs = plt.subplots(2,1, figsize=(16,7))
axs[0].plot(a[0])
axs[1].plot(a[1])
axs[0].set_title(data_path[0])
axs[1].set_title(data_path[1])

plt.show()
```
